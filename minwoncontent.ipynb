```python
html = driver.page_source  #현재 화면의 html을 모두 불러옴
soup = BeautifulSoup(html, 'html.parser')  #불러오는 정보를 걸러주는 도구
```

```python
minwon_dates = [ date.get_text() for date in soup.find_all('td', attrs={'class':'td-date'})] # 민원 작성일
minwon_date = []
for date in minwon_dates:
    if re.search(r'\d+.\d+.\d+.', date) != None: # 작성일 정제
        date = re.sub("-",".",date)
        minwon_date.append(date)
```

```python
df=pd.DataFrame()
for page in range(24):
    for i in range(1,21):
        path='//*[@id="dataSetTb"]/table/tbody/tr[{}]/td[2]/a'.format(i)
        driver.find_element(By.XPATH,path).click()
        time.sleep(1)

        dhtml = driver.page_source  #현재 화면의 html을 모두 불러옴
        dsoup = BeautifulSoup(dhtml, 'html.parser')  #불러오는 정보를 걸러주는 도구

        minwon_content = dsoup.select("table > tbody > tr> td ")
        content=[]
        for i in minwon_content:
            con = i.text.strip()
            content.append(con)

        df = df.append({'page':page+1,'title':content[1],'content':content[7],'answer':content[-1],'depart':content[8]}, ignore_index=True)

        #목록버튼
        #driver.find_element(By.XPATH,'/html/body/main/div/div[3]/a').click()
        driver.back()  #뒤로가기
        time.sleep(1)
    
    if page > 100: 
        element = driver.find_element(By.XPATH,'/html/body/main/div/form/div[2]/a[11]')
    else: element = driver.find_element(By.XPATH,'/html/body/main/div/form/div[2]/a[12]')
    driver.execute_script("arguments[0].click();", element)
    time.sleep(2)
content_df77=df
```
